```python
#Nomor 1
import pandas as pd

# Memuat data CSV
df = pd.read_csv('Movie_classification.csv')

# Menampilkan data
df.head()
```

```python
#Nomor  2

# Mengecek duplikat data
df.duplicated()
```

```python
#Nomor 3
# Menghitung persentase nilai null di setiap kolom
null_percentage = df.isnull().mean() * 100

# Menampilkan hasil persentase nilai null
null_percentage
```

```python
#Nomor 4
# Menghapus baris yang memiliki missing values
df_dropped_rows = df.dropna()

# Menampilkan data setelah baris yang mengandung missing values dihapu
df_dropped_rows
```

```python
# Menghapus kolom yang memiliki missing values
df_dropped_columns = df.dropna(axis=1)

# Menampilkan data setelah kolom yang mengandung missing values dihapus
df_dropped_columns
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Step 1: Check the data types of each column
print(df.dtypes)

# Step 2: Convert columns to numeric where applicable
# You can specify the columns you want to convert or apply to the entire DataFrame
for column in df.columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Step 3: Fill missing values with the mean for numeric columns
df_mean_imputed = df.fillna(df.mean())

# Display the DataFrame after mean imputation
print(df_mean_imputed)
```

```python
# Mengisi missing values dengan median untuk setiap kolom numerik
df_median_imputed = df.fillna(df.median())

# Menampilkan data setelah imputasi dengan median
df_median_imputed
```

```python
# Mengisi missing values dengan modus (mode) untuk setiap kolom
# Karena mode bisa lebih dari satu nilai, kita ambil nilai pertama dari modus
df_mode_imputed = df.fillna(df.mode().iloc[0])

# Menampilkan data setelah imputasi dengan modus
df_mode_imputed

```

```python
#Nomor 5

# Menyimpan data ke file CSV
output_csv_path = 'Movie_classification.csv'
df.to_csv(output_csv_path, index=False)
print(f"Data telah disimpan ke {output_csv_path}")

# Menyimpan data ke file Excel
output_excel_path = 'Movie_classification.xlsx'
df.to_excel(output_excel_path, index=False)
print(f"Data telah disimpan ke {output_excel_path}")

```
