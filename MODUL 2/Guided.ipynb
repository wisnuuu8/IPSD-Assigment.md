```python
import pandas as pd

# Load the iris dataset (ensure this step is correct)
# iris = pd.read_csv('path_to_your_iris.csv')

# Check the original column names
print("Original column names:")
print(iris.columns)

# Rename columns to lowercase
iris.columns = [col.lower() for col in iris.columns]

# Now convert the 'sepal_length' column to numeric
iris['sepal_length'] = pd.to_numeric(iris['sepal_length'], errors='coerce')

# Display the transformed DataFrame
print(iris.head())
```

```python
pivot_table = iris.pivot_table(values='sepal_length', index='target', aggfunc='mean')

# Tampilkan pivot table
print(pivot_table)
```
